# Twitter and Price - 10 minutes

## Import

```python
import pandas as pd
import os
```

## BTC Data
### Read

```python
os.chdir("/Users/stevelee/Downloads")
df = pd.read_csv('BTCUSD_transactions.csv')
df['Date'] = pd.to_datetime(df['timestamp'], unit='s')
df['Weighted Price'] = df['amount'] * df['price']
df = df.set_index('Date')
df.head()
```

### Resample

```python
df_btc = pd.DataFrame()
df_btc['Weighted Price Sum'] = df['Weighted Price'].resample('10min').sum()
df_btc['Amount'] = df['amount'].resample('10min').sum()
df_btc['Price'] = df_btc['Weighted Price Sum'] / df_btc['Amount']
df_btc.head()
```

## Twitter Data

### Read

```python
os.chdir('/Users/stevelee/Documents/Research/DigitalAsset_Analysis/data')
df_twitter = pd.read_csv('bitcoinTweets_big.csv', encoding = "ISO-8859-1", error_bad_lines=False)
df_twitter['Date'] = pd.to_datetime(df_twitter['Date'], format='%Y/%m/%d %H:%M:%S', errors='coerce')
df_twitter = df_twitter.set_index('Date').drop_duplicates(subset='TweetId', keep=False)
df_twitter.head()
```

### Resample 

```python
df_twitterSummary = pd.DataFrame()
df_twitterSummary['Retweets'] = df_twitter['Retweets'].resample('10min').sum()
df_twitterSummary['Favorites'] = df_twitter['Favorites'].resample('10min').sum()
df_twitterSummary['NumTweets'] = df_twitter['Favorites'].resample('10min').count()
df_twitterSummary
```

## Merge

```python
df_all = df_btc.merge(df_twitterSummary, how='outer', left_index=True, right_index=True).dropna(subset=['NumTweets'])
df_all
```

```python
df_all.to_csv('TwitterPriceData_tenMin.csv')
```
