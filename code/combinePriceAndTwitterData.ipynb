## Imports

```python
import os

import numpy as np
import pandas as pd
```

## Constants

```python
FOLDER_READ_TWITTER = "/Users/stevelee/Dropbox/TwitterData"
FOLDER_READ_PRICE = "/Users/stevelee/Dropbox/Spring2018_Coursework/ECON8812/project/data"
FOLDER_WRITE = "C:\\Users\\smlee\\Dropbox\\ECON8812\\project\\"

AUG = "bitcoin_Aug17.csv"
SEPT = "bitcoin_Sept17.csv"
OCT = "bitcoin_Oct17.csv"
NOV = "bitcoin_Nov17.csv"
DEC = "bitcoin_Dec17.csv"
JAN = "bitcoin_Jan18.csv"

PRICE_DATA = "coinData_hourly(Aug17-May18).csv"
```

## Read

```python
os.chdir(FOLDER_READ_TWITTER)

aug = pd.read_csv(AUG).set_index("date").sort_index()
sept = pd.read_csv(SEPT).set_index("date").sort_index()
october = pd.read_csv(OCT).set_index("date").sort_index()
nov = pd.read_csv(NOV).set_index("date").sort_index()
dec = pd.read_csv(DEC).set_index("date").sort_index()
jan = pd.read_csv(JAN).set_index("date").sort_index()

frames = [aug, sept, october, nov, dec, jan]

# drop text and username info for now
twitterData = pd.concat(frames).reset_index()
print(twitterData.head())
```

```python
os.chdir(FOLDER_READ_PRICE)
prices = pd.read_csv(PRICE_DATA).set_index('Time').sort_index().drop("Unnamed: 0",axis=1)

print(prices.head())
```

## Define function to adjust time lags in twitter data

```python
def adjustTwitterDateTime(twitterData):
    
    # set desired final series info
    finalTimeSeries = prices.index
    timePeriod = finalTimeSeries[1] - finalTimeSeries[0]
    
    # first valid time is one hour before the first price observation
    firstValidTime = finalTimeSeries[1] - timePeriod
    
    tweetDates = twitterData['date']
    twitterData['adjustedDate'] = None
    
    lastIndex = len(twitterData)
    indexCounter = 0
    for time in finalTimeSeries:
        print(time)
        
        # move counter to a valid starting time
        while (tweetDates[indexCounter] < firstValidTime):
            indexCounter += 1
        
        while (indexCounter < lastIndex and tweetDates[indexCounter] <= time):
            twitterData['adjustedDate'].iloc[indexCounter] = time
            indexCounter += 1
        
        if indexCounter >= lastIndex:
            return twitterData

t = adjustTwitterDateTime(twitterData)
print(t)
```

```python
grouped = t.groupby('adjustedDate')

times = []
retweets = []
favorites = []

for date, frame in grouped:
    _retweets = frame['retweets'].sum()
    _favorites = frame['favorites'].sum()
    
    times.append(date)
    retweets.append(_retweets)
    favorites.append(_favorites)
```

```python
twitterSums = pd.DataFrame({'time':times, 'retweets':retweets, 'favorites':favorites})
twitterSums = twitterSums.set_index('time')
twitterSums.head()
```

```python
masterData = prices.merge(twitterSums, how='outer', left_index=True, right_index=True)
masterData
```

## Save to CSV

```python
masterData.to_csv("priceWithTwitterData.csv")
```
